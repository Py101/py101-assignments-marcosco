```python
import matplotlib.pyplot as plt

from library_task_3 import read_json, words
from wordcloud import WordCloud

battles = read_json('battles.json')

print(words(battles))
# Generate a word cloud image
wordcloud = WordCloud().generate(' '.join(words(battles)))

# Display the generated image:
# the matplotlib way:

plt.imshow(wordcloud, interpolation='bilinear', aspect='auto')
plt.axis("off")

plt.show()
```
